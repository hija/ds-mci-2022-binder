```python
from sklearn.datasets import make_blobs, make_circles, make_moons
from sklearn.cluster import KMeans, DBSCAN

import seaborn as sns
```

```python
X, y = make_blobs(n_samples=100, centers=2, n_features=2, random_state=170)
```

```python
x1 = X[:, 0]
x2 = X[:, 1]
km = KMeans(n_clusters=2)
km.fit(X)
sns.scatterplot(x=x1, y=x2, hue=km.predict(X))
```

--> <strong>How do the algorithm cluster?</strong>

```python
# Example: make_moons with KMeans
X, y = make_moons(n_samples=100, random_state=170)
x1 = X[:, 0]
x2 = X[:, 1]


km = KMeans(n_clusters=2)
km.fit(X)
sns.scatterplot(x=x1, y=x2, hue=km.predict(X))
```

```python
# Example: make_moons with DBSCAN
X, y = make_moons(n_samples=1000, random_state=170)
x1 = X[:, 0]
x2 = X[:, 1]


dbscan = DBSCAN(eps=0.1)
sns.scatterplot(x=x1, y=x2, hue=dbscan.fit_predict(X))
```

```python
# Example: make_circles with DBSCAN
X, y = make_circles(n_samples=1000, random_state=170)
x1 = X[:, 0]
x2 = X[:, 1]


dbscan = DBSCAN(eps=0.15)
sns.scatterplot(x=x1, y=x2, hue=dbscan.fit_predict(X))
```

```python
# Example: make_circles with KMeans
X, y = make_circles(n_samples=1000, random_state=170)
x1 = X[:, 0]
x2 = X[:, 1]

km = KMeans(n_clusters=2)
km.fit(X)
sns.scatterplot(x=x1, y=x2, hue=km.predict(X))
```
