```python
import pandas as pd

# Your code here:
influencer_df = pd.read_csv("data/Top_Influencers.csv")
```

```python
influencer_df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.scatterplot(data=influencer_df, x='Followers', y='Avg. Likes')
plt.xscale('log')
plt.yscale('log')
```

```python
sns.histplot(data=influencer_df, x='Influence Score')
```

```python
sns.violinplot(data=influencer_df, x='Influence Score')
```

```python
sns.boxplot(data=influencer_df, x='Influence Score')
```

```python
sns.swarmplot(data=influencer_df, x='Influence Score')
```
